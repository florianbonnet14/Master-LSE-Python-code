We first need to establish the daily recommended values. To do so we first establish a list of references and then will cross-check:

    Precise portions of Fruits and Vegetables and also other food groups as well
- US Source 1:https://health.gov/sites/default/files/2019-09/2015-2020_Dietary_Guidelines.pdf
    Page 40-45 has cup-equivelent intakes
    p80 onwards recommendation in cup equivalent based on eenergy intake
- US Source 2: https://ods.od.nih.gov/HealthInformation/Dietary_Reference_Intakes.aspx#:~:text=Recommended%20Dietary%20Allowance%20(RDA)%3A,assumed%20to%20ensure%20nutritional%20adequacy.
    not the main categories but the details of the nutirents
- US Source 3: https://www.choosemyplate.gov/eathealthy
    Get the main groups in terms of cup equivalents
- EU Source 1: https://www.efsa.europa.eu/en/interactive-pages/drvs

```python
ADR_list_value = [400,400,30,150,200,28,55,8.5,850,90]
```

```python
ADR = pd.DataFrame(list(zip(ADR_list, ADR_list_value)), 
               columns =['Category', 'Daily Value(g,g,g,g,g,g,g,mg,mg,mg)']) 
```

```python
ADR
```

First we will pull the Recipes in order for Vegeteblaes, Fruits and Grains info, using the list of ingrdients

```python
List_ingredients_cat_weight = pd.read_csv('Recipes/ExcelFiles_distributions/recipes_ingredients_weights_flattened.csv')
pivot=List_ingredients_cat_weight.pivot_table('weight', 'recipe', 'category_refined', aggfunc=sum)
pivot = pivot.fillna(0)
Recipes_ADR_vfg = pivot[['dried fruits','fruits','fruit juice','vegetable','grains']]
Recipes_ADR_vfg.insert(0,'recipes',pivot.index)
Recipes_ADR_vfg = Recipes_ADR_vfg.reset_index()
```

```python
Recipes_ADR_vfg.head()
```

Then we will do the other nutrients part using the list of recipes and nutrients

```python
List_recipes_nutrients = pd.read_csv('Recipes/ExcelFiles_distributions/recipe_nutrients.csv')
Recipes_ADR_fpicvc = List_recipes_nutrients[['recipe','Protein','Fiber','Iron','Calcium','Vitamin C']]
```

Combining all in one

```python
Recipes_ADR = Recipes_ADR_vfg.join(Recipes_ADR_fpicvc.set_index('recipe'), on='recipes')
Recipes_ADR = Recipes_ADR.set_index('recipe').join(servings_recipes.set_index('recipe'))
Recipes_ADR['vegetable'] = Recipes_ADR['vegetable']/Recipes_ADR['persons']
Recipes_ADR['fruits'] = Recipes_ADR['fruits']/Recipes_ADR['persons']
Recipes_ADR['fruit juice'] = Recipes_ADR['fruit juice']/Recipes_ADR['persons']
Recipes_ADR['grains'] = Recipes_ADR['grains']/Recipes_ADR['persons']
Recipes_ADR['Fiber'] = Recipes_ADR['Fiber']/Recipes_ADR['persons']
Recipes_ADR['Protein'] = Recipes_ADR['Protein']/Recipes_ADR['persons']
Recipes_ADR['Iron'] = Recipes_ADR['Iron']/Recipes_ADR['persons']
Recipes_ADR['dried fruits'] = Recipes_ADR['dried fruits']/Recipes_ADR['persons']
Recipes_ADR['Calcium'] = Recipes_ADR['Calcium']/Recipes_ADR['persons']
Recipes_ADR['Vitamin C'] = Recipes_ADR['Vitamin C']/Recipes_ADR['persons']
Recipes_ADR_tmp= Recipes_ADR [['vegetable','fruits','dried fruits', 'fruit juice', 'grains', 'Fiber', 'Protein', 'Iron', 'Calcium', 'Vitamin C']]
Recipes_ADR_tmp.index = Recipes_ADR[['recipes']]      #getting rid of the two recipes columns ot be able to do divisions
ADRT = ADR[['Daily Value(g,g,g,g,g,g,g,mg,mg,mg)']].T    #trasnposing ADR, keeping only the vlues and putting names in columns names
ADRT.columns = ADR_list
tmp = Recipes_ADR_tmp.div(ADRT.iloc[0], axis = 'columns')
Recipes_ADR.index = Recipes_ADR[['recipes']]
tmp.insert(0,'recipe',Recipes_ADR[['recipes']])
tmp.reset_index()
Recipes_ADR_final=tmp.fillna(0)
Recipes_ADR_final['TotalFruits']= Recipes_ADR_final['dried fruits'] + Recipes_ADR_final['fruits'] + Recipes_ADR_final['fruit juice']
Recipes_ADR_final
```

```python
Recipes_ADR_final.to_csv('Recipes/ExcelFiles_distributions/recipes_ADR_final.csv')
```

Moderation uses simple derivation from nutrients but 2 elements need to be converted from g to KCal as thy are expressed in % of Energy. To do so we use : http://www.fao.org/3/y5022e/y5022e04.htm:

### Empty Calorie food

Nutrient density is the ratio of nutrients to energy in a certain amount of food consumed compared with the recommended levels.
*Example:*
 Vitamin C nutrient density of 100 g of apple consumed by a 20-y-old woman is 2.83
 - Vitamin C (100g apple) = 5.7 mg
 - Energy (100g apple) = 247 kJ
 - Recommended Vitamin C intake = 75 mg
 - Recommended Energy intake = 9211 kJ
 - Nutrient Ratio = (5.7 mg/75 mg)/(247 kJ/9211 kJ)
 
In DQI-I, if the sum of nutrient densities across nutrients examined in a food is < 1 then the food is considered to be an empty calorie food.
Then, if the energy supplied by empty calorie food is > 10% of total energy intake per day, owest score is assigned.

Since we do not have access to the components per ingredients but only at receipe level, we will compute a nutrient density score at recipe level using the following nutrients, leveraging insights from https://academic.oup.com/jn/article/139/8/1549/4670510:

# Overall Balance

Overall Balance is computing Macronutrient ratio (carbohydrate, protein, fat) and Fatty Acid Ratio.

Macronutrient Ratio will ne computed once recipes have been selected. For test with only 1 recipes, we will emasure departure from mean ie only on one recipe level.
- 55-65/10-15/15-25 --> 6
- 52-68/9-16/13-27 --> 4
- 50-70/8-17/12-30 --> 2
- Others --> 0
To compute this ratio of energy, we need to use conversion tables from http://www.fao.org/3/y5022e/y5022e04.htm:
 - Carbohydrates: 17 kcal/g
 - Protein:4 kcal/g 
 - Fat: 9 kcal/g

Fatty Acid Ratio will be computed from the follwing rule:
- P/S = 1-1.5 / M/S = 1-1.5 -->4
- P/S = 0-8-1.7 / M/S = 0.8-1.7 -->2
- Others --> 0

```python
List_nutrients = pd.read_csv('Recipes/ExcelFiles_distributions/recipe_nutrients.csv')
```

```python
List_nutrients.columns
```

```python
List_balance = List_nutrients[['recipe','Energy','Carbs','Protein', 'Fat', 'Monounsaturated','Polyunsaturated','Saturated']]
```

```python
List_balance['Carbs kJ'] = List_balance['Carbs']*17
List_balance['Protein kJ'] = List_balance['Protein']*4
List_balance['Fat kJ'] = List_balance['Fat']*9
List_balance['Monounaturated kJ'] = List_balance['Monounsaturated']*9
List_balance['Polyunsaturated kJ'] = List_balance['Polyunsaturated']*9
List_balance['Saturated kJ'] = List_balance['Saturated']*9
```

```python
List_balance = List_balance.set_index('recipe').join(servings_recipes.set_index('recipe'))
```

```python
List_balance['Energy'] = List_balance['Energy']/List_balance['persons']
List_balance['Carbs'] = List_balance['Carbs']/List_balance['persons']
List_balance['Protein'] = List_balance['Protein']/List_balance['persons']
List_balance['Fat'] = List_balance['Fat']/List_balance['persons']
List_balance['Monounsaturated'] = List_balance['Monounsaturated']/List_balance['persons']
List_balance['Polyunsaturated'] = List_balance['Polyunsaturated']/List_balance['persons']
List_balance['Saturated'] = List_balance['Saturated']/List_balance['persons']
List_balance['Carbs kJ'] = List_balance['Carbs kJ']/List_balance['persons']
List_balance['Protein kJ'] = List_balance['Protein kJ']/List_balance['persons']
List_balance['Fat kJ'] = List_balance['Fat kJ']/List_balance['persons']
List_balance['Monounaturated kJ'] = List_balance['Monounaturated kJ']/List_balance['persons']
List_balance['Polyunsaturated kJ'] = List_balance['Polyunsaturated kJ']/List_balance['persons']
List_balance['Saturated kJ'] = List_balance['Saturated kJ']/List_balance['persons']
```
